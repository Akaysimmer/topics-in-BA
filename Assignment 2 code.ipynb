# Data cleaning
#now we shall clean the data

```python
# Dropping all the non-numerical columns
numdf = df_train.drop(['name', 'release_date', 'artists', 't_name0', 't_name1', 't_name2'], axis = 1)
```

```python
numdf.head()
```

```python
#check number of missing values
numdf.isnull().sum()
```

```python
#We need to come up with a strategy to deal with the remaining missing values in other columns.
#The choice we made is replacing missing values in other columns using column means:
numdf = numdf.fillna(numdf.mean())
```

```python
#Let us check if there are indeed no missing values anymore:
numdf.isnull().sum()
```

```python
#finding the mean across t_dur0, t_dur1, t_dur2 and creating a colum t_dur for the average
numdf['dur'] = numdf.iloc[:, 1:4].mean(axis=1)
```

```python
numdf.head()
```

```python
#finding the mean across t_dance0, t_dance1, t_dance2 and creating a column t_dance for the average
numdf['dance'] = numdf.iloc[:, 4:7].mean(axis=1)
```

```python
numdf.head()
```

```python
#finding the mean across the rest of the features and creating a new column for the average
numdf['energy'] = numdf.iloc[:, 7:10].mean(axis=1)
numdf['key'] = numdf.iloc[:, 10:13].mean(axis=1)
numdf['mode'] = numdf.iloc[:, 13:16].mean(axis=1)
numdf['speech'] = numdf.iloc[:, 16:19].mean(axis=1)
numdf['acous'] = numdf.iloc[:, 19:22].mean(axis=1)
numdf['ins'] = numdf.iloc[:, 22:25].mean(axis=1)
numdf['live'] = numdf.iloc[:, 25:28].mean(axis=1)
numdf['val'] = numdf.iloc[:, 28:31].mean(axis=1)
numdf['tempo'] = numdf.iloc[:, 31:34].mean(axis=1)
numdf['sig'] = numdf.iloc[:, 34:37].mean(axis=1)
```
